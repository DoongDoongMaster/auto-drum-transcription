```python
def check_condition(obj):
    # 'KK' 키를 제외한 나머지 키에서 [1] 값을 가지는 키의 개수를 카운트합니다.
    count = sum(1 for key, value in obj.items() if key != 'KK' and key != 'CH' and value == [1])
    
    # [1] 값을 가지는 키의 개수가 3개 이상이면 False를 반환합니다.
    return count < 3

# 테스트를 위한 입력 객체
input_obj = {
    "CC": [1],
    "OH": [1],
    "CH": [1],
    "TT": [0],
    "SD": [0],
    "KK": [0]
}

# 함수 호출 및 결과 출력
print(check_condition(input_obj))
```
